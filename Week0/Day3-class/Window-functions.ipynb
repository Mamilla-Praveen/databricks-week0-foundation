### 1.Assign a unique row number to all employees based on salary (highest first).

```python
%sql
select row_number() over(order by salary desc) as row_num,*
from employees
```

### 2. Assign row numbers to employees within each department based on salary descending.

```python
%sql
select row_number() over(partition by department order by salary desc) as row_num,*
from employees
```

### 3. Assign row numbers based on employee joining date (latest first). 

```python
%sql
select row_number() over(order by join_date desc) as row_number,* from employees
```

### 4. Assign row numbers within each department based on earliest joining date.

```python
%sql
select row_number() over(partition by department order by join_date desc) as row_number,* from employees
```

### 5. Assign row numbers to orders based on order date. 

```python
%sql
select row_number() over(order by order_date desc) as row_number,* from orders
```

### 6. Assign row numbers to orders within each city based on order amount (highest first).

```python
%sql
select row_number() over(partition by city order by order_amount desc) as row_num,*
from orders
```

### 7. Assign row numbers to employees based on salary (lowest first).

```python
%sql
select row_number() over(order by salary asc) as row_num,* from employees
```

### 8. Assign row numbers within department for employees based on name alphabetically.

```python
%sql
select row_number() over(partition by department order by emp_name) as row_number,* from employees
```
