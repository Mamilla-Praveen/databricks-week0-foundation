```python
data = [
    (1, "A", 5000),
    (2, "B", 6000),
    (3, "C", 7000)
]

columns = ["id", "name", "salary"]

df = spark.createDataFrame(data, columns)
df.display()
```

### write as Delta Table

```python
df.write.format("delta") \
    .mode("overwrite") \
    .save("/Volumes/workspace/default/my-volume")
```

### Read Delta Table

```python
df= spark.read.format("delta").load("/Volumes/workspace/default/my-volume")
df.display()
```

```python
new_data = [(4,"D",8000)]
new_df = spark.createDataFrame(new_data,columns)

new_df.write.format("delta")\
    .mode("append")\
        .save("/Volumes/workspace/default/my-volume")
```

```python
new_data = spark.read.format("delta")\
    .load("/Volumes/workspace/default/my-volume")
new_data.display()
```
